```python
import yfinance as yf
import pandas as pd

# download daily SPY data (2010–2025)
data = yf.download("SPY", start="2010-01-01", end="2025-01-01", auto_adjust=True)

print(data.shape)
data.head()
```

```python
data.to_csv("../data/raw/spy.csv")
```

```python
df = pd.read_csv("../data/raw/spy.csv", header=[0,1], index_col=0, parse_dates=True)

# Flatten column names (drop the top level like "Price")
df.columns = [col[0] for col in df.columns.to_flat_index()]

df.head()
```
