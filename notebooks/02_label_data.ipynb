```python
import yfinance as yf
import pandas as pd

df = pd.read_csv("../data/raw/spy.csv", header=[0,1], index_col=0, parse_dates=True)

df["return_1d"] = df["Close"].shift(-1) / df["Close"] - 1

df.head()
```

```python
df = df.dropna()

df.tail()
```

```python
df["y"] = (df["return_1d"] > 0).astype(int)

df.head()
```

```python
import os

output_dir = os.path.join("..", "data", "processed")
os.makedirs(output_dir, exist_ok=True)

df.to_csv(os.path.join(output_dir, "spy_labeled.csv"), index=False)
```
